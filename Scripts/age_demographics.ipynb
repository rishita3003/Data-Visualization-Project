```python
import pandas as pd

# Read the CSV data
df = pd.read_csv("./Data/yearwise_AgeDemographics.csv")
df
```

```python
# Extract age groups (all columns except Year and FTAs)
age_groups = df.columns[2:]

# Create a new DataFrame with age_labels as the index
age_data = pd.DataFrame({'age_labels': age_groups})
age_groups, age_data
```

```python
# Add year columns with actual visitor counts
for _, row in df.iterrows():
    year = int(row['Year'])
    ftas = row['FTAs']
    
    # Calculate actual visitor counts for each age group based on percentage
    for age_group in age_groups:
        percentage = row[age_group]
        visitor_count = int(round((percentage / 100) * ftas))
        age_data.loc[age_data['age_labels'] == age_group, str(year)] = visitor_count

age_data
```

```python
# Save the transformed data
age_data.to_csv('transformed_tourism_age_data.csv', index=False)
```
